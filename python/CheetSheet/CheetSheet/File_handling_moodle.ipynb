Exercise 1 – Display the following menu to the user:
Create a new file
Display the file
Add a new item to the file
Make a selection to 1, 2 or 3
Ask the user to enter 1, 2 or 3. If they select anything other than 1, 2 and 3, it should display a suitable error message. If they select 1, ask the user to enter a school subject and save it to a new file called “Subjects.txt”. It should overwrite any existing file with a new file. If they select 2, display the contents of “Subjects.txt” file. If they select 3, ask the user to enter a new subject and save it to the file and then display the entire content of the file. Run the programme several times to test the options.


```python
# Exercise 1: Menu to Create, Display, and Add School Subjects to a file

# File name to store subjects
filename = "Subjects.txt"

while True:
    # Display menu
    print("\n--- School Subjects Menu ---")
    print("1. Create a new file")     # Overwrites old file
    print("2. Display the file")      # Show contents
    print("3. Add a new item to file")# Append new subject
    
    choice = input("Enter 1, 2 or 3: ")  # Ask user choice

    if choice == "1":
        # Option 1: create a new file and overwrite any old file
        subject = input("Enter a school subject: ")
        with open(filename, "w") as f:  # 'w' mode overwrites
            f.write(subject + "\n")     # Add subject + newline
        print("New file created and subject added.")

    elif choice == "2":
        # Option 2: display contents of the file
        try:
            with open(filename, "r") as f:  # Open in read mode
                print("\n--- Subjects ---")
                for line in f:             # Loop through each line
                    print(line.strip())    # Remove newline characters
        except FileNotFoundError:
            print("File does not exist. Please create it first.")  # Error if file not found

    elif choice == "3":
        # Option 3: append new subject and display full file
        new_subject = input("Enter a new subject: ")
        with open(filename, "a") as f:      # 'a' mode appends
            f.write(new_subject + "\n")
        print("Subject added. Current file content:")
        with open(filename, "r") as f:
            for line in f:
                print(line.strip())

    else:
        print("Invalid selection! Enter 1, 2 or 3.")  # Handle wrong input
```

Exercise 2 -  Create a simple maths quiz that will ask the user for their name and then generate two random questions. Store their name, the questions that were asked, their answers and their final score in a .csv file. Whenever the programme is run it should add to the csv file, and not overwrite anything.


```python
import csv
import random

# Ask user for their name
name = input("Enter your name: ")

questions = []  # To store question, user answer, correct answer
score = 0       # Initialize score

# Generate 2 random math questions
for i in range(2):
    a = random.randint(1, 20)           # Random number 1
    b = random.randint(1, 20)           # Random number 2
    op = random.choice(["+", "-"])       # Random operation
    q = f"{a} {op} {b}"                 # Question string
    answer = eval(q)                     # Compute correct answer
    user_ans = input(f"Question {i+1}: {q} = ")  # Ask user
    try:
        if float(user_ans) == answer:    # Compare answers
            score += 1
    except:
        pass
    questions.append((q, user_ans, answer))  # Save details

# Append results to CSV file (does not overwrite)
with open("MathQuiz.csv", "a", newline="") as file:
    writer = csv.writer(file)
    # Write: name, Q1, user_ans, correct_ans, Q2, user_ans, correct_ans, score
    writer.writerow([
        name,
        questions[0][0], questions[0][1], questions[0][2],
        questions[1][0], questions[1][1], questions[1][2],
        score
    ])

print(f"\n{name}, your score is {score}/2")
print("Quiz recorded in MathQuiz.csv")
```

Exercise 3 - (Function and .csv files) Create the following menu:
Add to file
View all records
Quit programme
Enter the number of your selection:
If the user selects 1, allow them to add to a file called Salaries.csv which will store their name and salary. If they select 2 it should display all records in the Salaries.csv file. If they select 3 it should stop the programme. If they select an incorrect option, they should see an error message. They should keep returning to the menu until they select option 3.


```python
import csv

filename = "Salaries.csv"  # CSV file to store data

while True:
    # Display menu
    print("\n--- Salaries Menu ---")
    print("1. Add to file")
    print("2. View all records")
    print("3. Quit programme")

    choice = input("Enter your choice: ")

    if choice == "1":
        # Add record
        name = input("Enter name: ")
        salary = input("Enter salary: ")
        with open(filename, "a", newline="") as f:  # Append mode
            writer = csv.writer(f)
            writer.writerow([name, salary])          # Write row to CSV
        print("Record added.")

    elif choice == "2":
        # View all records
        try:
            with open(filename, "r") as f:
                reader = csv.reader(f)
                print("\n--- All Salaries ---")
                for row in reader:
                    print(row)
        except FileNotFoundError:
            print("File does not exist. Please add records first.")

    elif choice == "3":
        print("Exiting program...")
        break

    else:
        print("Invalid choice! Enter 1, 2 or 3.")
```

Exercise 4 – (function and csv) In Python, it is not technically possible to directly delete a record from a .csv file. Instead, you need to save the file to a temporary list in Python, make the changes to the list and then overwrite the original file with the temporary list.
Change the previous programme to allow you to do this. Your menu should now look like this:
Add to file
View all records
Delete a record
Quit programme

```python
import csv

filename = "Salaries.csv"

while True:
    # Display menu
    print("\n--- Salaries Menu ---")
    print("1. Add to file")
    print("2. View all records")
    print("3. Delete a record")
    print("4. Quit programme")

    choice = input("Enter your choice: ")

    if choice == "1":
        # Add record
        name = input("Enter name: ")
        salary = input("Enter salary: ")
        with open(filename, "a", newline="") as f:
            writer = csv.writer(f)
            writer.writerow([name, salary])
        print("Record added.")

    elif choice == "2":
        # View all records
        try:
            with open(filename, "r") as f:
                reader = csv.reader(f)
                print("\n--- All Salaries ---")
                for row in reader:
                    print(row)
        except FileNotFoundError:
            print("File does not exist.")

    elif choice == "3":
        # Delete record by name
        del_name = input("Enter name to delete: ")
        try:
            with open(filename, "r") as f:
                rows = list(csv.reader(f))   # Read all into list
            with open(filename, "w", newline="") as f:
                writer = csv.writer(f)
                deleted = False
                for row in rows:
                    if row[0] != del_name:   # Keep rows not matching name
                        writer.writerow(row)
                    else:
                        deleted = True
                if deleted:
                    print("Record deleted.")
                else:
                    print("Name not found.")
        except FileNotFoundError:
            print("File does not exist.")

    elif choice == "4":
        print("Exiting program...")
        break

    else:
        print("Invalid choice! Enter 1-4.")
```

Exercise 5 - Sum a Collection of Numbers
Create a program that sums all of the numbers entered by the user while ignoring any input that is not a valid number. Your program should display the current sum after each number is entered. It should display an appropriate message after each non-numeric input, and then continue to sum any additional numbers entered by the user. Exit the program when the user enters a blank line. Ensures that your program works correctly for both integer and floating-point numbers.

```python
# Exercise 5: Sum numbers entered by user
# Ignore invalid inputs
total = 0

print("Enter numbers to sum. Press Enter (blank) to finish.")

while True:
    num = input("Enter a number: ")
    if num == "":       # Blank line ends the loop
        break
    try:
        number = float(num)   # Convert input to float
        total += number       # Add to total
        print(f"Current sum: {total}")
    except ValueError:
        print("Invalid input! Please enter a valid number.")  # Ignore non-numbers

print(f"Final sum: {total}")
```
